### Extracting card dedicated capacity and convert all into Gb like 1 TB = 1024 Gb

````python
df["capacity_gb"] =df["card"].str.split(",").str.get(1).str.strip().str.split("upto").str.get(-1).str.strip()


def extract(value):
        if pd.isna(value):
          return 0
        parts = str(value).split()
        if "TB" in str(value):
        # Extract number and convert TB to GB
          number = int(parts[0])
          return number * 1024
        elif "GB" in str(value):
        # Extract just the number
         return int(parts[0])
        else:
          return 0

    
df["capacity_gb"] = df["capacity_gb"].apply(extract)
````
