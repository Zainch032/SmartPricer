```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
df  = pd.read_excel("data\Smart-phone.xlsx")
```

```python

df["Company"] = df["model"].apply(lambda x : x.split(" ")[0])
```

```python
df["Company"].replace({"OPPO":"Oppo"},inplace=True)
df["Company"].value_counts()
```

```python
df1  = df.copy()
```

```python
df[df["price"]==df["price"].max()]
```

```python
df1  = df1[df1["price"]>=3400]
```

```python
df1["price"].describe()
```

```python
df1["processor"].value_counts()
```

```python
df1 = df1.reset_index()
processor = set((786,974,979,985))
filtered_df = df1[df1["index"].isin(processor)]  
filtered_df
```

```python
df1.drop([786,974,979,985,529],inplace=True)
```

```python
print(df1[df1["Company"].isin(["Nokia"])& (df1["price"] < 4000)])
df1.drop([988],inplace=True)
```

#### Extract the phones which have inches in battery by wrong

```python

temp_df = df1[df1["battery"].apply(lambda x : x.split(" ")[1].strip()).str.contains("inch")]
```

```python
x  = temp_df.iloc[:,7:-1].shift(1,axis=1).values
```

```python
x_index = temp_df.index
df1.loc[temp_df.index,temp_df.columns[7:-1]] = x 
df1.isnull().sum()
```

```python
df1[df1["Company"].str.contains("Nokia")]["model"].to_dict()
```

#### Removing features phones

```python
df1.drop([963, 978, 982, 989],inplace = True)
```

#### Phone which have issue in camera wrong data -> Display

```python
temp_df = df1[df1["camera"].str.contains("Display")]
temp_df.shape
```

```python
df1.loc[temp_df.index,'camera'] = temp_df["card"].values
```

```python
temp_df2 = df1[df1["card"].str.contains("MP")]
df1.loc[temp_df2.index,'card'] = 'Memory Card Not Supported'
df1["card"].value_counts()
```

```python
temp_df3 = df1[~df1["card"].str.contains("Card")]
x  = temp_df3.iloc[:,10:-1].shift(1,axis=1)
df1.loc[temp_df3.index,'os'] = x
df1.loc[temp_df3.index,'card'] = "Memory Card Not Supported"
```

```python
df1['os'].value_counts()
```

#### All card value is perfect

```python
df1['card'].value_counts()
```

#### Wrong card value in os column due to shifting

```python
temp_df4 = df1[df1['os'].str.contains("Card")]
len(temp_df4)
```

### Right shift these card values in card coloums from os

```python
x  = temp_df4.iloc[:,10:-1].shift(-1,axis=1)
df1.loc[temp_df4.index,'card']  = x
z = df1[df1["os"].str.contains("Bluetooth")]
df1.loc[z.index,'card']  = np.nan
df1.loc[temp_df4.index,'os'] = np.nan
```

```python

print(df1['os'].value_counts())
df1['card'].value_counts()
```
